```python
import numpy as np
import pandas as pd
```

```python
# Load data CartPole-v1
with open("./out/cartpole-v1/cartpole_v1_ppo_15000.csv") as f:
    df_cartpole_v1 = pd.read_csv(f, index_col='timesteps', header=0)
```

```python
# Load data CartPole-v1k
with open("./out/cartpole-v1k/cartpole_v1k_ppo_15000.csv") as f:
    df_cartpole_v1k = pd.read_csv(f, index_col='timesteps', header=0)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=[7,3])

plt.plot(df_cartpole_v1.index, df_cartpole_v1['mean_reward'], label='cartpole_v1', zorder=100, color='blue')
plt.scatter(df_cartpole_v1.index, df_cartpole_v1['mean_reward'], zorder=100, color='darkblue')

plt.plot(df_cartpole_v1k.index, df_cartpole_v1k['mean_reward'], label='cartpole_v1k', zorder=100, color='orange')
plt.scatter(df_cartpole_v1k.index, df_cartpole_v1k['mean_reward'], zorder=100, color='red')

plt.title('Mean reward')
plt.xlabel('mean reward')
plt.ylabel('timesteps')
plt.legend()
plt.xlim((0, 15000))
plt.ylim((0, 1050))
plt.show()
```

```python
fig = plt.figure(figsize=[7,3])

plt.plot(df_cartpole_v1.index, df_cartpole_v1['std_reward'], label='cartpole_v1', zorder=100, color='blue')
plt.scatter(df_cartpole_v1.index, df_cartpole_v1['std_reward'], zorder=100, color='darkblue')

plt.plot(df_cartpole_v1k.index, df_cartpole_v1k['std_reward'], label='cartpole_v1k', zorder=100, color='orange')
plt.scatter(df_cartpole_v1k.index, df_cartpole_v1k['std_reward'], zorder=100, color='red')

plt.title('std reward')
plt.xlabel('std reward')
plt.ylabel('timesteps')
plt.legend()
plt.xlim((0, 15000))
plt.ylim((0, 350))
plt.show()
```
