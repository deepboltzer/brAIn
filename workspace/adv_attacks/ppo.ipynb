```python
import numpy as np
import gym
```

```python
from stable_baselines3 import PPO
from stable_baselines3.ppo.policies import MlpPolicy
```

```python
# Train model
env = gym.make('CartPole-v1')

model = PPO(MlpPolicy, env, verbose=1)
model.learn(100000)
```

```python
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.monitor import Monitor
```

```python
# Evaluate model - for info on why the env is wrapped with Monitor check the evaluate_policy function
mean_reward, std_reward = evaluate_policy(model, Monitor(env), n_eval_episodes=100)
print(f"mean_reward : {mean_reward:.2f} +/- {std_reward:.2f}")
```

```python
# Save model
save_dir = "./models/ppo"
model.save(save_dir + f"/cartpole-v1/cartpole_v1_ppo_100k")
```
