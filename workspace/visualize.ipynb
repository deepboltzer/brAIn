```python
import numpy as np
import pandas as pd
```

```python
# Load data CartPole-v1
with open("./out/cartpole-v1/cartpole_v1_ppo_15000.csv") as f:
    df_cartpole_v1 = pd.read_csv(f, index_col='timesteps', header=0)
```

```python
# Load data CartPole-v1k
with open("./out/cartpole-v1k/cartpole_v1k_ppo_15000.csv") as f:
    df_cartpole_v1k = pd.read_csv(f, index_col='timesteps', header=0)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot(x, y, label=None, xlabel=None, ylabel=None, title=None, legend=True, xlim=(0, 15000), ylim=(0, 550), color='blue'):
    """Visualize data as plot.
    :param x: plot y versus x
    :param y: plot y versus x
    :param label: label of the plot
    :param xlabel: label of x axis
    :param ylabel: label of y axis
    :param title: figure title
    :param legend: show legend
    :param xlim: interval for x
    :param ylim: interval for y
    """
    fig = plt.figure(figsize=[7,3])
    plt.plot(x, y, label=label, zorder=100, color=color)
    plt.scatter(x, y, zorder=100, color='black')
    if title:
        plt.title(title)
    if xlabel:
        plt.xlabel(xlabel)
    if ylabel:
        plt.ylabel(ylabel)
    if legend:
        plt.legend()
    plt.xlim(xlim)
    plt.ylim(ylim)
```

```python
# Plot CartPole-v1 model
plot(
    df_cartpole_v1.index, 
    df_cartpole_v1['mean_reward'], 
    label='mean reward', 
    xlabel='timesteps', 
    ylabel='mean rewards', 
    title='CartPole-v1 (PPO)'
    )
    
plot(
    df_cartpole_v1.index, 
    df_cartpole_v1['std_reward'], 
    label='std reward', 
    xlabel='timesteps', 
    ylabel='std rewards', 
    title='CartPole-v1 (PPO)', 
    ylim=(0, 200))

# Plot CartPole-v1k model
plot(
    df_cartpole_v1k.index, 
    df_cartpole_v1k['mean_reward'], 
    label='mean reward', 
    xlabel='timesteps', 
    ylabel='mean rewards', 
    title='CartPole-v1k (PPO)',
    ylim=(0, 1050),
    color='red'
    )

plot(
    df_cartpole_v1k.index, 
    df_cartpole_v1k['std_reward'], 
    label='std reward', 
    xlabel='timesteps', 
    ylabel='std rewards', 
    title='CartPole-v1k (PPO)',
    ylim=(0, 300),
    color='red'
    )
```
